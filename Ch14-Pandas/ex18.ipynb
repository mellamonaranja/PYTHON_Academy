```python
def peak_to_peak_ratio(x):
    return x.max() / x.min()
```

```python
iris.species
```

```python
iris.groupby(iris.species)
```

```python
result = iris.groupby(iris.species).agg(peak_to_peak_ratio )
```

```python
result = iris.groupby(iris.species).agg(peak_to_peak_ratio() )
```

```python
result
```

```python
def max3(df):
    return df.sort_values(by='petal_length')[:3]
```

```python
iris.sort_values(by='petal_length')[:3]
```

```python
result = iris.groupby(iris.species).apply(max3)
```

```python
result
```

```python
def q3cut(s):
    return pd.qcut(s, 3, labels=['소', '중', '대'])
```

```python
iris.groupby(iris.species)['petal_length'].transform(q3cut)
```

```python
iris['petal_length_class'] = iris.groupby(iris.species)['petal_length'].transform(q3cut)
```

```python
iris
```

```python
result = iris[['petal_length', 'petal_length_class']].tail(10)
```

```python
result
```
