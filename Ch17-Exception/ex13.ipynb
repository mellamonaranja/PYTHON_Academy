```python
class SquareSeq:
    def __init__(self, n):
        self.n = n
        
    # a[2] -> __getitem__(self, k) X
    def __getitem__(self, k):
        return k*k
    
        
    # len() -> __len__(self)
    def __len__(self):
        return self.n
```

```python
class SquareSeq:
    def __init__(self, n):
        self.n = n
        
    # a[2] -> __getitem__(self, k) X
    def __getitem__(self, k):
        
        if k >= self.n or k < 0:
            raise IndexError
        
        return k*k
    
        
    # len() -> __len__(self)
    def __len__(self):
        return self.n
```

```python
s = SquareSeq(10)
```

```python
type(s)
```

```python
s[2]
```

```python
s[4]
```

```python
s[20]
```

```python
s[20000000]
```

```python
for x in s:
    print(x, end= " ")
```
